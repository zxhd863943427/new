## 多回归分析及拟合判断

&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160; **描述性分析到此为止，接下来就到了最激动人心的时候了！我们将利用手上现有的数据建立对于金价的预测模型！**

```python
# 导入绘图函数
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))

    ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
    ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.xlabel('Price (in dollars)')
    plt.ylabel('Proportion of Cars')

    plt.show()
    plt.close()
```

- ###  **进行多变量二阶回归建模**

```python
# 导入多变量计算模块
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 导入线性拟合模块
from sklearn.linear_model import LinearRegression
```

```python
Z = df_data[['America GDP','America RGDP','China GDP','China RGDP','world GDP']]
```

```python
#构建多变量高级特征提取模型
pr = PolynomialFeatures(degree=2)
```

```python
Z_pr = pr.fit_transform(Z)
```

```python
# 构建线性拟合模型
lm = LinearRegression()
```

```python
lm.fit(Z_pr,df_data['gold value'])
```

- ###  **观察测试数据与原数据的拟合程度-1**

```python
Z_test=lm.predict(Z_pr)
```

```python
DistributionPlot(Z_test, df_data['gold value'], 'x', 'gold value', 'a test')
```

```python
lm.score(Z_pr,df_data['gold value'])
```

#### <kbd>分析</kbd>：从拟合图像来看，拟合效果相当优秀，R方评估显示拟合程度达到了97.44%，超出了我的预计

- ###  **观察测试数据与原数据的拟合程度-2**

&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160;&#160; **我们需要观测一下模型与原数据相关性和残差分布，确定是否有系统误差**

```python
test = pd.DataFrame()
```

```python
test['predict'] = lm.predict(Z_pr)
```

```python
test['gold value']=df_data['gold value']
```

```python
sns.regplot('predict','gold value',data=test)
```

```python
width = 5
height = 5
plt.figure(figsize=(width, height))
sns.residplot(test['predict'], test['gold value'])
plt.show()
```

#### <kbd>分析</kbd>：从相关性绘图来看，拟合效果优秀，从残差图来看，误差随机均匀分布在坐标轴上下，可以证明误差基本为随机误差，而并非拟合失误导致都系统误差
