```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# @Time    : 2018/10/22 9:17
# @Author  : Sa.Song
# @Desc    : jieba分词与wordcloud提取词云
# @File    : jieba_wordcloud.py
# @Software: PyCharm

import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import jieba.analyse


with open(r'test.txt', 'r') as f:  # 本地文本文档
    text = f.read()
with open(r'stop.txt', 'r', encoding='utf-8') as f:  # 本地的停用词表
    tingyongci = f.read()

path = r'C:\Users\admin\Pictures\STARS-19071-207.jpg'  # 本地的图片模板
font = Fonts  # 本地的字体（中文分词要用）

text = text.replace('，','')
text = text.replace('1','')
text = text.replace('、','')
text = text.replace('；','')
text = text.replace('.','')
text = text.replace('。','')
str_list = jieba.cut(text, HMM=True) #使用精准模式来分词

'''加载停用词表并去掉停用词'''
outstr = ''
for word in str_list:
    if word not in tingyongci:
        if word != '\t':
            outstr += word
            outstr += ' '
```

```python
img = Image.open(path)  # 打开图片
img_array = np.array(img)  # 将图片转换为数组
x,y = np.ogrid[:3000,:3000]
mask = (x-1500) ** 2 + (y-1500) ** 2 > 1300 ** 2
mask = 255 * mask.astype(int)
wc = WordCloud(
    background_color='black',
    mask=mask,
    font_path=font,
    # mode='RGBA',
    max_words=100  # 提取的关键词的最大个数
)
wc.generate_from_text(outstr)  # 根据切好的词来绘制词云图，自动提取关键词
# wc.generate_from_frequencies(new_outstr)  # 根据给定的关键词和词频（字典格式）来画词云图
#plt.figure()  # 创建一个画图窗口
#plt.imshow(wc, interpolation='bilinear')
#plt.axis('off')
plt.show()
wc.to_file(r'C:\Users\admin\Pictures\b.png')
```

- 导入模块

```python
import numpy as np
import jieba 
import wordcloud
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

- 导入文件和停用词列表

```python
txt_path = r'test.txt'
stop_path = r'stop.txt'
with open(txt_path,'r') as f:
    file = f.read()
with open(stop_path,'r',encoding='utf-8') as f:
    stop = f.read()
```

```python
stop=set(stop)
```

- 分词

```python
str1=jieba.cut(file,HMM=True)
```

```python
# 保存分词列表
str_list =[]
for word in str1:
    str_list.append(word)
```

```python
outstr=''
for word in str_list:
    if word not in stop:
        if word != '\u3000':
            outstr += word
            outstr += ' '
```

- 建立词云

```python
x,y = np.ogrid[:3000,:3000]
mask = (x-1500)**2+(y-1500)**2 > 1300**2
mask = 255*mask.astype(int)
```

```python
Fonts = r'C:\Windows\Fonts\STFANGSO.TTF'
wc = wordcloud.WordCloud(background_color='white',
                        font_path=Fonts,
                         mask=mask,
                        max_words=500,
                        #width=3000,
                        #height=1500,
                        max_font_size=500)
wc.generate_from_text(outstr)
```
